# 1. 손글씨 데이터 셋


```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.keys()
```

```python
data = digits['data']
data.shape
```

```python
data[0]
```

```python
target = digits['target']
target.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 5, figsize=(14, 8))

for i, ax in enumerate(axes.flatten()): #flatten다차원배열을 1차원 배열로 보여준다
  ax.imshow(data[i].reshape([8,8]), cmap='gray')
  ax.set_title(target[i])
```

```python
fig, axes = plt.subplots(2, 5, figsize=(14, 8))
print(fig)
print(axes)
```

# 2. 정규화


```python
data[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled = scaler.fit_transform(data) # 이 객체를 사용하여 데이터를 fit하고 변환(transform)합니다. fit 과정에서는 데이터의 최대값과 최소값을 계산하고, transform 과정에서는 이 정보를 바탕으로 데이터를 0과 1 사이의 값으로 변환합니다.
scaled[0] # 스케일링된 데이터의 첫 번째 샘플을 출력합니다.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled, target, test_size=0.2, random_state=10)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# 3. Support Vector Machine(SVM)
* 두 클래스로부터 최대한 멀리 떨어져 있는 결정 경계를 찾는 분류기로 특정 조건을 만족하는 동시에 클래스를 분류하는 것을 목표로함

<img src="https://i.imgur.com/l1NQNvD.png" width="700px">

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(y_test[0], y_pred[0])
plt.imshow(X_test[0].reshape(8,8))
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 5, figsize=(14, 8))

for i, ax in enumerate(axes.flatten()):
  ax.imshow(X_test[i].reshape(8,8), cmap='gray')
  ax.set_title(f'Label: {y_test[i]}, Pred: {y_pred[i]}')
```
