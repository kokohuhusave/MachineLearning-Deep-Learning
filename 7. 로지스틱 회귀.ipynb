# 3. 혼종 행렬(confusion matrix)
* 정밀도 와 재현율(민감도)를 활용하는 평가용 지수
* 

TN(8869)     FP(0)
FN(862)      TP(1)
* TN(승진하지 못했는데 승진 하지 못했다고 예측)
* FN(승진하지 못했는데 승진 했다고 예측)
* FP(승진 했는데 승진 못했다고 예측)
* TP(승진 했는데 승진 했다고 예측)

````python
confusion_matrix(y_test, pred)
````

````python
sns.heatmap(confusion_matrix(y_test, pred), annot=True, cmap='Blues')
#연과관계 확인
````

### 3-1. 정밀도
* TP / (TP + FP)
* 무조건 양성으로 판단해서 계산하는 방법
* 실제 1인 것중에 얼마 만큼을 제대로 맞췄는가?

### 3-2. 재현율
* TP / (TP + FN)
* 정확하게 감지한 양성 샘플의 비율
* 1이라고 예측한 것들중 얼마나 1이 있는가
* 민감도 또는 TPR(True Positive Rate)라 불리기도 한다.

### 3-3. f1 score
* 정밀도와 재현율의 조화평균을 나타내는 지표

$$2*\frac{정밀도 * 재현율}{정밀도 + 재현율}=\frac{TP}{TP+\frac{FN+FP}{2}}$$

```
정밀도  재현율  산술평균  조합평균
0.4      0.6      0.5       0.48
0.3      0.7      0.5       0.42
0.5      0.5      0.5       0.5
```


### Confusion Matrix에 따른 **정확도** 재정의

앞에서 언급한 정확도는 다음과 같다.
$$
\begin{align*}
정확도(Accuracy) =
{\operatorname{맞게예측한결과건수}\over\operatorname{전체예측데이터건수}}
\end{align*}
$$

이 정확도를 Confusion Matrix 분류를 이용하여 재 정의 하면 다음과 같다.
$$
\begin{align*}
정확도(Accuracy) =
{\operatorname{TN + TP}\over\operatorname{TN + FP + FN + TP}}
\end{align*}
$$


### 정밀도(Precision)과 재현율(Recall) 정의

---
**정밀도(Precision)**는 참(Positive)으로 예측한 대상(FP + TP)중에 실제 참인값(TP)의 비율을 이야기 한다.
$$
\begin{align*}
정밀도(Precision) =
{\operatorname{TP}\over\operatorname{FP + TP}}
\end{align*}
$$

다르게 이야기를 하면, 이 학습된 인공지능모델이 참이라고 예측한게 얼마나 정확한지에 대한 척도이다.

---


**재현율(Recall)**은 전체 예측 대상중 참(FN + TP)인 데이터중에 참으로 인식해서 맞춘(TP)비율을 이야기 한다.

$$
\begin{align*}
재현율(Recall) =
{\operatorname{TP}\over\operatorname{FN + TP}}
\end{align*}
$$

다르게 이야기를 하면, 실제 데이터중 전체 참(Positive)중에 얼마나 많은 참(TP)를 찝어 낼수 있는가에 대한 척도이다. 민감도(Sensitivity) 혹은 TPR(True Positive Rate)라 불리기도 한다.

---

# 4.교차 검증(Cross Validation)
* train_test_split에서 발생하는 데이터의 섞임에 따라 성능이 좌우되는 문제를 해결하기 위한 기술
* K-Fold 교차검증을 가장 많이 사용                                                                                                                                                                        

````python
from sklearn.model_selection import KFold
````

````python
kf = KFold(n_splits=5)
````

````python
kf
````

````python
hr_df
````

````python
for train_index, test_index in kf.split(range(len(hr_df))):
  print(train_index, test_index)
  print(len(train_index), len(test_index))
````

````python
kf = KFold(n_splits=5, random_state=10, shuffle=True)
````

````python
for train_index, test_index in kf.split(range(len(hr_df))):
  print(train_index, test_index)
  print(len(train_index), len(test_index))
````

````python
acc_list = []
for train_index, test_index in kf.split(range(len(hr_df))):
    X = hr_df.drop('is_promoted', axis=1)
    y = hr_df['is_promoted']
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y.iloc[train_index]
    y_test = y.iloc[test_index]
    lr = LogisticRegression()
    lr.fit(X_train, y_train)
    pred = lr.predict(X_test)
    acc_list.append(accuracy_score(y_test, pred))
````
