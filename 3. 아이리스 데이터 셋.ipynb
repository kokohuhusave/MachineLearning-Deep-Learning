```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
iris
```

```python
print(iris['DESCR'])
```

- sepal length in cm: 꽃받침의 길이
- sepal width in cm: 꽃받침의 너비
- petal length in cm: 꽃잎의 길이
- petal width in cm: 꽃잎의 너비


```python
data = iris['data']
data
```

```python
target = iris['target']
target
```

```python
feature_names = iris['feature_names']
feature_names
```

```python
import pandas as pd
df_iris = pd.DataFrame(data,columns=feature_names)
df_iris.head(3)
```

```python
df_iris['target'] = target
df_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train_test_split(독립변수, 종속변슈, 테스트 사이즈,시드값)
# 
X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target', 1), 
                                                    df_iris['target'], 
                                                    test_size=0.2, 
                                                    random_state=10)
X_train
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print('정답률: ',accuracy_score(y_test, y_pred))
```
