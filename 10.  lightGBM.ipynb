# 2. lightGBM(LGBM)
* 트리기반 학습 알고리즘인 fradient boosting 방식의 프레임 워크
* 의사결정나무, 랜덤포레스트는 균형 트리 분할(Level wise)방식이라면, LGBM은 리프 중심 트리 분할(leaf wise)
* GBM(Gradient Boosting): 모델 1을 통해 y를 예츨하고, 모델2에 데이터를 넣어 y를 예측, 모델3에 넣어 y를 예측
* 학습하는데 걸리는 시간이 적음
* 메모리 사용량이 상대적으로 적은 편
* 적은 데이터셋을 사용할 경우 과적합 가능성이 매우 큼(일반적으로 데이터가 10000개 이상은 사용해야 함)

```python
from lightgbm import LGBMClassifier
```

```python
base_model = LGBMClassifier(random_state=10)
```

```python
base_model.fit(X_train, y_train)
```

```python
pred1 = base_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report, roc_auc_score
```

```python
accuracy_score(y_test, pred1)
```

```python
confusion_matrix(y_test, pred1)
```

```python
print(classification_report(y_test, pred1))
```

```python
proba1 = base_model.predict_proba(X_test)
proba1
```

```python
9.15719979e-01
```

```python
roc_auc_score(y_test, proba1, multi_class='ovr') #LGBM의 값
```

# 3. RamdomizedSearchCV
* 분류기를 결정하고 해당 분류기의 최적의 하이퍼 파라미터를 찾기 위하 방법
* 튜닝하고 싶은 파라미터를 지정하여 파타미터 값의 범위를 설정하고, n_iter값을 설정하여 해당 수 만큼 Random하게 조합하여 반복 적용

```python
# n_estimators: 반복 수행하는 트리의 갯수(기본값: 100), 값을 크게 지정하면 학습시간도 오래걸리며, 과적합이 발생할 수 있음
# max_depth: 트리의 최대 깊이(기본값: -1)은 끝까지
# num_leaves:  각각의 LightGBM 트리에 있는 최대 리프 노드 수를 지정하는 매개변수입니다. 리프 노드는 트리의 가장 끝에 위치한 노드로, 트리에서 예측을 수행하는 단말 노드입니다.
# leaning_rate: 학습률(기본값: 0.1) 얼마 만큼 이동해서 측정해야 최적의 값이 나올까
params = {
    'n_estimators':[100,300,500,1000],
    'max_depth':[-1,30,50,100],
    'num_leaves':[5,10,20,40],
    'learning_rate':[0.01,0.05,0.1,0.5]
}
```

```python
lgbm = LGBMClassifier(random_state=10)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rand_lgbm = RandomizedSearchCV(lgbm, params, n_iter=30, random_state=10)
```

```python
rand_lgbm.fit(X_train, y_train)
```

```python
rand_lgbm.cv_results_
```

```python
rand_lgbm.best_params_
```

```python
lgbm = LGBMClassifier(random_state=10, num_leaves=10,n_estimators=1000,max_depth=100,learning_rate=0.01 )
```

```python
lgbm.fit(X_train, y_train)
```

```python
proba = lgbm.predict_proba(X_test)
```

```python
roc_auc_score(y_test, proba, multi_class='ovr')
```
