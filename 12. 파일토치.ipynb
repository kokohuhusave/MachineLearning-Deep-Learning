# 1. 파일 토치
* 텐서플로우와 함꼐 머신러닝, 딥러닝에서 가장 널리 사용되는 프레임워크

```python
import torch
print(torch.__version__)
```

### 1-1. 스칼라(Scalar)
* 하나의 상수를 의미


```python
var1 = torch.tensor([1])
```

```python
type(var1)
```

```python
var2 = torch.tensor([6.5])
```

```python
# 두 스칼라의 덧셈연산
print(var1 + var2)
print(var1 - var2)
print(var1 * var2)
print(var1 / var2)
```

### 1-2. 벡터(Vector)
* 상수가 두 개 이상 나열 되었을 경우


```python
vector1 = torch.tensor([1,2,3])
vector1
```

```python
vector2 = torch.tensor([10,20,30])
```

```python
#두 벡터의 사칙 연산
print(vector1 + vector2)
print(vector1 - vector2)
print(vector1 / vector2)
print(vector1 * vector2)
```

### 1-3. 행렬(Matrix)
* 2개 이상의 벡터값을 가지고 만들어진 값으로 행과 열의 개념을 가진 숫자의 나열

```python
matrix1 = torch.tensor([[1,2],[3,4]])
print(matrix1)
```

```python
matrix2 = torch.tensor([[7,8],[9,10]])
print(matrix2)
```

```python
#두 행렬의 사칙 연산
print(matrix1 + matrix2)
print(matrix1 - matrix2)
print(matrix1 * matrix2)
print(matrix1 / matrix2)
```

### 1-4. 텐서(Tensor)
* 다수의 행렬이 모이면 텐서라고 부름
* 배열이나 행렬과 매우 유사한 특수한 자료구조
* 파일토치는 텐서를 사용해서 모델의 입력과 출력, 모델의 매개변수들을 처리 시용됨

화면 캡처 2023-06-15 093249.jpg![image.png]()https://miro.medium.com/max/875/1*jRyyMAhS_NZxqyv3EoLCvg.png

```python
tensor1 = torch.tensor([[[1,2],[3,4]],[[5,6] ,[7,8]]])
print(tensor1)
```

```python
tensor2 = torch.tensor([[[9,10],[11,12]],[[13,14] ,[15,16]]])
print(tensor2)
```

```python
print(tensor1 + tensor2)
print(tensor1 - tensor2)
print(tensor1 / tensor2)
print(tensor1 * tensor2)
```

```python
print(torch.add(tensor1, tensor2))
print(torch.subtract(tensor1, tensor2))
print(torch.multiply(tensor1, tensor2))
print(torch.divide(tensor1, tensor2))
print(torch.matmul(tensor1, tensor2)) # 행렬곱
```

```python
print(tensor1.add_(tensor2)) # tensor1에 결과를 다시저장 모든 사칙 연산자에 _를 붙여 사용할 수있다
print(tensor1)
```

# 2. 텐서의 변환

```python
data = [[1,2], [3,4]]
print(data)
```

```python
x_data = torch.tensor(data)
print(x_data)
```

```python
import numpy as np
```

```python
np_array = np.array(data)
np_array
```

```python
x_np_1 = torch.tensor(np_array)
x_np_1
```

```python
x_np_1[0, 0] = 100
print(x_np_1)
print(np_array)
```

```python
x_np_2 = torch.as_tensor(np_array)# ndarray와 동일한 메모리 주소를 가리키는 뷰를 만드는 함수
print(x_np_2)
x_np_2[0, 0] = 200  # 기존 메모리 주소의 ndarray값을 변경하게 됨
print(x_np_2)
print(np_array) #메모리를 같이 쓴다
```

```python
x_np_3 = torch.from_numpy(np_array)
print(x_np_3)
x_np_3[0, 0] = 400 # 기존 메모리의 주소의 ndarray값을 변경하게 됨
print(x_np_3)
print(np_array)
```

```python
np_again = x_np_1.numpy() #다시 numpy ndarray로 변경
print(np_again, type(np_again))
```

# 3.파이토치 주요 함수


```python
a = torch.ones(2,3)
print(a)# 1로 채운다
```

```python
b = torch.zeros(2,3)
print(b)#0으로 채운다
```

```python
c = torch.full((2, 3), 10) #2행 3열로 10으로 채운다
print(c)
```

```python
f = torch.arange(10) # 0부터 n까지의 숫자를 넣는다
print(f)
```

```python
g = torch.rand(2, 3) # 0부터 1사이의 숫자를 뽑아 넣는다
print(g)
```

```python
i = torch.arange(16).reshape(2, 2, 4) #2개 matrix 2행 4열 (0 , 1, 2)번 자리로 생각
print(i, i.shape)

j = i.transpose(1,2) # 1번 2번을 가져온다 2, 4, 2가 된다
print(j) # 2개의
# tensor([[[ 0,  4],
#          [ 1,  5],
#          [ 2,  6],
#          [ 3,  7]],

#         [[ 8, 12],
#          [ 9, 13],
#          [10, 14],
#          [11, 15]]])
```

```python
k = i.permute((2, 0, 1)) #i 자체가 (2, 2, 4)  2, 2, 4 -> 4, 2, 2
print(k,k.shape) # 0 , 1, 2자리 확인
```

# 4. GPU사용하기
* 코랩에서 device변경하는 방법
  * 싱단 메뉴 -> 런타임 -> 런타임 유형변경 -> 하드웨어 가속기를 GPU로 변경 -> 저장 -> 다시시작


```python
tensor = torch.rand(3,4)
print(f'shape: {tensor.shape}')
print(f'dtype: {tensor.dtype}')
print(f'device: {tensor.device}')
```

```python
tensor = tensor.reshape(4, 3)
tensor = tensor.int()
if torch.cuda.is_available(): #cuda 토치에 GPU를 가리킴
  print('GPU를 사용할 수 있음')
  tensor = tensor.to('cuda')# GPU로 보내 재저장

print(f'shape: {tensor.shape}')
print(f'dtype: {tensor.dtype}')
print(f'device: {tensor.device}')
```
